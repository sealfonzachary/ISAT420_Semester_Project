<p style="color:black"> Ghost forests are expanding in the chesapeke bay as coastal forests are exposed to increasing saltwater intrusion. To better understand this problem, long term environmental data is needed to examine how salinity patterns relate to ghost forest encroachment over time</p>

<h3 style="color:darkgreen;text-align:left">Problem Background and Introduction</h3>


<p>As mentioned, ghosts forests are expanding in the Chesapeake Bay region due to environmental changes that stem from rising sea levels. Ghosts forests occur when forests near the coast are killed because of a higher exposure to salt water, which leaves behind standing dead "ghost" trees. This process reduces biodiversity and change wildlife habitat.</p>
<p>One of the largest concerns about ghost forests is the ecosystem succession after the ghost forests die. On one hand, there is a change that the forests becomes a wetland biome, such as a saltwater or brackish marsh. These ecosystems are actually significantly more productive than the forest systems they replace. However, there is also the chance for monocultural invasive species, notably common reeds (Phragmites) to take over. Common reeds are aggressively invasive, and tend to choke out competition. This, of course, leads to a biodiversity-poor ecosystem and hurts the health of the Chesapeake Bay. </p>
<p> This project will examine how patterns caused by climate change in the Chesapeake Bay are related to ghost forest encroachment over time. </p>
