```python
import requests
import pandas as pd

url = "http://datahub.chesapeakebay.net/api.JSON/WaterQuality/WaterQuality/01-01-1990/12-31-2020/0,1/2/12/HUC8/20/23"

headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)

data = response.json()

df = pd.DataFrame(data)

df.to_csv("cbp_salinity_api.csv", index=False)

print(df.head())
```
