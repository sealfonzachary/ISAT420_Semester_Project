```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

station = "8635750"  # Lewisetta, VA
params = {
    "begin_date": "19940101", 
    "end_date": "20241231",
    "station": station,
    "product": "monthly_mean",
    "datum": "MSL",
    "units": "metric",
    "time_zone": "lst_ldt",
    "format": "json"
}
response = requests.get("https://api.tidesandcurrents.noaa.gov/api/prod/datagetter", params=params)
data = response.json()

# Create the DataFrame from the 'data' list
df_trend = pd.DataFrame(data['data'])

# 1. Create a proper timestamp by joining Year and Month
df_trend['timestamp'] = pd.to_datetime(df_trend['year'] + "-" + df_trend['month'])

# 2. Extract the MSL values (converting from text to numbers)
# We use errors='coerce' just in case there's a missing month
df_trend['sea_level'] = pd.to_numeric(df_trend['MSL'], errors='coerce')

# 3. Clean up the index and remove empty rows
df_trend.set_index('timestamp', inplace=True)
df_trend = df_trend[['sea_level']].dropna()

print(f"Data processed for {len(df_trend)} months.")
df_trend.head()

plt.figure(figsize=(12, 6))

# Plot the raw monthly data
plt.plot(df_trend.index, df_trend['sea_level'], label='Monthly Mean', alpha=0.4, color='gray')

# Plot a 12-month rolling average to show the trend clearly
plt.plot(df_trend.index, df_trend['sea_level'].rolling(window=12).mean(), label='12-Month Average', color='red')

plt.title(f"30-Year Relative Sea Level Rise at {station} (Lewisetta, VA)")
plt.ylabel("Sea Level (Meters, MSL)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('30_Year_Relative_Sea_Level_Rise_at_Lewisetta_NOAA_Station.png', format='png')
plt.show()
```

```python
# 1. Define the future date
future_date = pd.Timestamp('2050-01-01')
future_x = future_date.toordinal()

# 2. Predict the sea level for 2050 using the slope and intercept from your model
# Formula: y = mx + b
predicted_msl_2050 = (slope * future_x) + intercept

# 3. Calculate the total rise from the last recorded data point
latest_date = df_trend.index[-1]
latest_msl = df_trend['sea_level'].iloc[-1]
total_rise_mm = (predicted_msl_2050 - latest_msl) * 1000

print(f"Predicted Sea Level (MSL) in Jan 2050: {predicted_msl_2050:.3f} meters")
print(f"Estimated rise from {latest_date.year} to 2050: {total_rise_mm:.1f} mm")

# --- NEW: Compute Trendline and R^2 ---
# Convert timestamps to ordinal numbers for regression (x-values)
x = df_trend.index.map(pd.Timestamp.toordinal)
y = df_trend['sea_level']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
r_squared = r_value**2

# Calculate the trendline values (y = mx + b)
trend_line = slope * x + intercept

print(f"Data processed for {len(df_trend)} months.")
print(f"Linear Trend: {slope * 365.25 * 1000:.2f} mm/year") # Convert to mm/year
print(f"R-squared: {r_squared:.4f}")

# --- Visualizing the Projection ---
plt.figure(figsize=(12, 6))

# Plot existing data
plt.plot(df_trend.index, df_trend['sea_level'], label='Historical Monthly Mean', alpha=0.3, color='gray')
plt.plot(df_trend.index, trend_line, color='green', label='Historical Trend')

# Create a projection line from the end of data to 2050
projection_dates = pd.date_range(start=df_trend.index[-1], end=future_date, freq='MS')
projection_x = projection_dates.map(pd.Timestamp.toordinal)
projection_y = slope * projection_x + intercept

plt.plot(projection_dates, projection_y, color='green', linestyle='--', label='2050 Projection')
plt.scatter(future_date, predicted_msl_2050, color='darkgreen', zorder=5)

plt.annotate(f"2050 Prediction: {predicted_msl_2050:.2f}m", 
             xy=(future_date, predicted_msl_2050), 
             xytext=(-140, 10), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='black'))

plt.title(f"Sea Level Rise Projection for {station} (Lewisetta, VA)")
plt.ylabel("Sea Level (Meters, MSL)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
